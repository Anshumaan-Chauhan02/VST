## Importing necessary libraries

```python
import os 
import pandas as pd
```

## Setting up OS

```python
direc_m = "/content/drive/MyDrive/Visual Story Telling/Dataset - Story Generation/Movies"
files_m = os.listdir(direc_m)

direc_b = "/content/drive/MyDrive/Visual Story Telling/Dataset - Story Generation/Books"
files_b = os.listdir(direc_b)
```

## Merging Movie sub-datasets

```python
merged_df = pd.DataFrame(columns= ['Wikipedia movie ID', 'Summary', 'Genre', 'Movie_Name', 'Characters', 'Relations'])
for file_name in files_m:
  path = direc_m + "/" + file_name
  local_df = pd.read_csv(path)
  merged_df = pd.concat([merged_df, local_df])
  merged_df.reset_index(drop=True, inplace= True)
merged_df.to_csv('/content/drive/MyDrive/Visual Story Telling/Dataset - Story Generation/Processed_Movie_Dataset',index = False)
```

## Merging Book sub-datasets

```python
merged_df = pd.DataFrame(columns= ['Wikipedia article ID', 'Book title', 'Book genres', 'Plot summary', 'Characters', 'Relations'])
for file_name in files_b:
  path = direc_b + "/" + file_name
  local_df = pd.read_csv(path)
  merged_df = pd.concat([merged_df, local_df])
  merged_df.reset_index(drop=True, inplace= True)
merged_df.to_csv("/content/drive/MyDrive/Visual Story Telling/Dataset - Story Generation/Processed_Book_Dataset", index = False)
```
